```python
flag = True
max_x = 0
max_y = 0
goal = 0
start = 0

def water(x, y):
    global flag
    global max_x
    global max_y
    global goal
    global start
    if(flag):
        max_x = int(input("Enter the maximum value of Jug A: "))
        max_y = int(input("Enter the maximum value of Jug B: "))
        goal = int(input("Enter the final amount of water you want in Jug A: "))
        start = int(input("Do you want to start with Jug A or Jug B?\nEnter 1 for Jug A\nEnter 2 for Jug B\n"))
        flag = False
    print("(" + str(x) + "," + str(y) + ")")
    if start == 2:
        if x == goal:
            # print("RULE 1")
            print("Mission Accomplished!")
            return 
        elif x == max_x:
            # print("RULE 2")
            water(0,y)
            #throwing out all the water of x and pouring all the water from y to x
        elif x == 0 and y != 0:
            # print("RULE 3")
            water(y, 0)
            #transfering all the water from y to x because x is empty
        elif y == goal:
            # print("RULE 4")
            water(0, y)
            #transfering the water from y to x 
        elif y < max_y:
            # print("RULE 5")
            water(x, max_y)
            #filling y completely 
        elif y < (max_x - x):
            # print("RULE 6")
            water(x+y, 0)
            #transfering y into x
        else:
            # print("RULE 7")
            water((max_x-x)+x, y-(max_x-x))
            # spliting the jugs by pouring water from Jug B to Jug A, so as to fill Jug A to the brim
    else:
        if x == goal:
            print("Mission Accomplished!")
            return 
        elif y == max_y:
            water(x, 0) 
            #throwing out all the water of y
        elif x == max_x:
            water(x-(max_y-y), y+(max_y-y))
            # spliting the jugs by pouring water from Jug A to Jug B
        elif x != 0 and y == 0:
            water(0, x)
            # transfering all the water from x to y because y is empty
        elif y == goal:
            water(0, y)
            #transfering the water from y to x 
        elif x < max_x:
            water(max_x, y)
            #filling x completely because its not filled to the brim
        elif y < (max_x - x):
            water(x+y, 0)
            #transfering y into x
        
        
print("Water Jug Problem")
water(0,0)
```
